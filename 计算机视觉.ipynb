## PIL：Python图像处理类库
PIL（Python Imaging Library Python，图像处理类库）提供了通用的图像处理功能，以及大量有用的基本图像操作，比如图像缩放、裁剪、旋转、颜色转换等。PIL 是免费的。

利用 PIL 中的函数，我们可以从大多数图像格式的文件中读取数据，然后写入最常见的图像格式文件中。PIL 中最重要的模块为 Image。要读取一幅图像，可以使用：

```python
# 加载PIL中的Image用于图像基本处理
from PIL import Image

# 图像显示
from matplotlib.pyplot import imshow

# pil_im执行后即可获得图像的二进制信息
pil_im = Image.open('empire.jpg')

# 显示图片
imshow(pil_im)
```

上述代码的返回值 pil_im 是一个 PIL 图像对象。


# 灰度处理

图像的颜色转换可以使用 convert() 方法来实现。要读取一幅图像，并将其转换成灰度图像，只需要加上 convert('L')，如下所示：

```python
pil_im = Image.open('empire.jpg').convert('L')
imshow(pil_im) 
```

# 图像裁剪

使用 crop() 方法可以从一幅图像中裁剪指定区域：

```python
box = (100,100,400,400)
im = Image.open('empire.jpg').crop(box)
imshow(im)
```

# 调整尺寸

```python
im = Image.open('empire.jpg').resize((128,128))
imshow(im)
```

# 图像旋转

```python
im = Image.open('empire.jpg').rotate(45)
imshow(im)
```

# 图像轮廓和直方图

```python
# 加载PIL中的Image用于图像基本处理
from PIL import Image

# 图像显示
from matplotlib.pyplot import imshow

from pylab import *

# pil_im执行后即可获得图像的二进制信息
pil_im = Image.open('empire.jpg')
 
# 读取图像到数组中
im = array(pil_im.convert('L'))
 
# 新建一个图像
figure()
# 不使用颜色信息
gray()
# 在原点的左上角显示轮廓图像
contour(im, origin='image')
axis('equal')
axis('off')

figure()
hist(im.flatten(),128)
```
